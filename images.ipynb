```python
# plot logistisc function
def adjust(y):
    return 100*(y - y.min()) / (y.max() - y.min())

def plot_logistic(x0, c):
    k = 0.8
    x = np.linspace(0, 10, 100)
    y = 1 - 1 / (1 + np.exp(-k * (x - x0)))
    plt.plot(x, adjust(y), color=c)
# plot_logistic(0.5, 10)
# with plt.xkcd(scale=1, randomness=1):
# set size of the plot
plt.figure(figsize=(8, 6))
with plt.xkcd():
    plot_logistic(8, 'crimson')
    plot_logistic(5, 'darkorange')
    plot_logistic(2, 'green')
plt.axhline(y=50, color='black', linestyle='--', linewidth=2)

plt.xticks(range(0, 11))
plt.yticks(np.linspace(0, 100, 5))
plt.legend(['High', 'Mid', 'Low', "Target"])
plt.ylim(0, 100)
plt.xlim(0, 10)
plt.title('Performance vs difficulty', fontsize=20)
plt.xlabel('Difficulty', fontsize=16)
plt.ylabel('Performance', fontsize=16)
# ignore the warning
# ignore all warnings from matplotlib


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the size of the matrix
# Create a matrix where values decline linearly along the x-axis
x = np.linspace(1, 0, 10)

# Create a matrix where values decline exponentially along the y-axis
y = np.exp(-np.linspace(0, 1, 10))

# Combine x and y to create the final matrix
data = 100*np.outer(y, x)
# with plt.xkcd():
# plt.figure()


Z = interpolate(data)
with plt.xkcd():
    fig, ax = plt.subplots(figsize=(8, 6))
    
    cbar = plt.colorbar(ax.matshow(Z, origin='lower', extent=(1, 10, 1, 10)))
    cbar.set_ticks(np.linspace(0, 100, 11))
    ax.xaxis.set_ticks_position('bottom')
    plt.title('Difficulty Grid', fontsize=20)
    plt.xlabel('p$_{death}$', fontsize=16)
    plt.ylabel('\u0394gap', fontsize=16)

    ax.set_xlim(1, 10)
    ax.set_ylim(1, 10)
    # Difficulty curve
    continuous = np.linspace(0, 99, 100)
    line = np.column_stack((continuous, continuous))  # Slope 1 line
    plt.plot(line[:, 1], line[:, 0], color='r')
plt.show()
```

## P(death)

```python
X = np.arange(1, 11)
Y = np.sqrt(X) / 100
P = (1-Y) ** 100
# set size
plt.figure(figsize=(8, 5))
plt.plot(X, P)
plt.ylim(0, 1)
plt.xticks(range(1, 11))
plt.xlim(1, 10)
plt.title('Probability of not dying randomly', fontsize=20)
plt.xlabel('p$_{death}$', fontsize=16)
plt.ylabel('Probability', fontsize=16)
plt.show()
```
